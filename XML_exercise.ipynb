1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
import pandas as pd
import xml.etree.ElementTree as ET
import json
tree = ET.parse('mondial.xml')
root = tree.getroot()
root
print(root.tag)
print(root.attrib)
print(len(root))
```

```python
my_dict = {'name':[],
          'infant_mortality':[]}
country= root[0].tag

for country in root:
    try:
        b = country.findtext('infant_mortality')
        my_dict['infant_mortality'].append(float(b))
        name_value = country[0].text
        my_dict['name'].append(name_value)
    except:
        pass
    
print(pd.DataFrame(my_dict).sort_values('infant_mortality', ascending=False).head(10))

```
